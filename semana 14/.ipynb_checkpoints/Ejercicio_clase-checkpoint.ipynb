# Uso y definici√≥n de funciones

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline #para activar graficos en el notebook
```

```python
x=np.linspace(-10,10,100)
x2=np.linspace(0,5,5)
y =x**2*np.cos(x)-3*x**3
y2=x2**2*np.cos(x2)-3*x2**3
plt.plot(x,y)
plt.plot(x2,y2)
```

```python
#como definir nuestras propias funciones
```

```python
#opcion para definir funciones 
fx=lambda x: x**2*np.cos(x)-3*x**3 #x es solo una etiqueta
```

```python
fx(x2)
```

```python
#opcion 2 para definir funciones
def fx2(x,y):
    z=x*y*np.cos(x/y)-3*x**3
    return z
```

```python
fx2
```

```python
plt.plot (x,fx2(x,y))
```

```python
#FUNCIONES CON ARGUMENTOS CTES A MENOS QUE EL USUARIO LOS CAMBIE
def recta (a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))
plt.plot(x,recta(x,3,-5))
plt.grid()
```

```python
def recta (a,m=1,b=0,label= 'x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
x=np.linspace(-2,2,20)
recta(x)
recta(x,-2,1, '-x+1')
plt.legend()
plt.grid()
```
