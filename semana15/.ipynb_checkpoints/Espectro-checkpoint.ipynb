```python
def Intensidad(v,T): #definicion de funcion
    v=v*c*100
    f1=h/c**2/1e-20 
    f2=h*v/(k*T)            #1 Jy =10^-26 W•m^-2•Hz
    return f1*v**3/(np.exp(f2)-1)
```

```python
plt.errorbar(x,Intensidad(x,2.5)) 
      
```
